```python
import time
import numpy as np
from math import exp, factorial, log
import matplotlib.pyplot as plt
from multiprocessing import Process, Queue
%pylab --no-import-all inline
```

```python
params = {'amin':75,
          'amax':90,
          'bmin':500,
          'bmax':600,
          'p1':0.1,
          'p2':0.01,
          'p3':0.3}

amin = params['amin']
amax = params['amax']
bmin = params['bmin']
bmax = params['bmax']
p1 = params['p1']
p2 = params['p2']
p3 = params['p3']
cmin = 0;
cmax = amax+bmax
dmin = 0
dmax = 2*(amax+bmax)

prob_threshold = 10e-4
cpu_count = multiprocessing.cpu_count()
```

```python
log_fact = np.zeros(cmax+1)
log_fact[0] = log(1)
for i in range(1, cmax+1):
    log_fact[i] = log_fact[i-1] + log(i)
```

```python
def estimate_max_value(probs):
    for i in range(len(probs))[::-1]:
        if (probs[i] > prob_threshold):
            return i
    return 0

a_distr = [bin(amax, p1, x) for x in xrange(amax+1)]
a_max_value = estimate_max_value(a_distr)
    
b_distr = [bin(bmax, p2, x) for x in xrange(bmax+1)]
b_max_value = estimate_max_value(b_distr)

c_max_value = a_max_value+b_max_value

f_stud_distr = [bin(c_max_value, p3, x) for x in xrange(c_max_value+1)]
f_stud_max_value = estimate_max_value(f_stud_distr)

d_max_value = c_max_value+f_stud_max_value

print "Сокращаем исходное множество допустимых параметров до тех, которые будут давать ненулевую вероятность c"
print "Аналогично сокращаем множество d"
print "a_max={0} b_max={1} c_max={2} f_stud_max={3} d_max={4}".format(a_max_value, b_max_value, c_max_value, f_stud_max_value, d_max_value)
```

```python
def uniform(min, max):
    len = max-min+1
    return np.ones(len)/len

def uniform_mean(min, max):
    return float(max+min)/2

def uniform_dev(min, max):
    n = max - min + 1;
    return float(n*n-1)/12

def poiss(n, lam):
    degree = -lam + n*log(lam) - log_fact[n]
    return exp(degree)

def bin(n, p, k):
    coeff = exp(log_fact[n]-log_fact[n-k]-log_fact[k])
    return coeff*pow(p,k)*pow(1-p,n-k)  
    
def mean(values, probs):
    return np.sum(values*probs)

def dev(values, probs, calculatedMean=None):
    if (calculatedMean == None):
        calculatedMean = mean(values, probs)
    return mean((values-calculatedMean)**2,probs)

def print_time_estimation(name, time):
    print "Оценка", name, "заняла", time, "мс"
    

def estimate_params(name, values, probs, verbose=True):
    mean_value = mean(values, probs)
    dev_value = dev(values, probs, mean_value)
    if (verbose):
        print "E({0})={1} D({0})={2}".format(name, mean_value, dev_value)
    return (mean_value, dev_value)

def find_nzero_ends(y):
    first_nzero = 0
    last_nzero = len(y)
    for i in range(len(y)):
        if (y[i] > prob_threshold):
            first_nzero = i
            break
    for i in range(len(y))[::-1]:
        if (y[i] > prob_threshold):
            last_nzero = i
            break
    return (first_nzero, last_nzero)
    
def plot_distr(name, x_label, x, y):
    plt.figure()
    (first, last) = find_nzero_ends(y)
    x = x[first:last+1]
    y = y[first:last+1]
    plt.plot(x, y, 'r')
    plt.xlabel(x_label)
    plt.ylabel('p')
    plt.title(name)
    plt.show()
    
def get_range(min_value, max_value, count):
    result = list()
    length = max_value - min_value
    part_len = length/count
    residue = length%count
    prev = min_value
    for i in xrange(count):
        next = prev+part_len
        if (residue > 0):
            next += 1
            residue -= 1
        result.append((prev, next))
        prev = next
    return result
    
```

Выведем необходимые формулы для пункта номер 2. <br>
$$p(c)=\sum_{a,b}p(c|a,b)$$
$$p(c|a)=\sum_bp(c|a,b)$$
$$p(c|b)=\sum_ap(c|a,b)$$
$$p(c|d)=\frac{p(d|c)p(c)}{p(d)}=\frac{p(d|c)p(c)}{\sum_cp(d|c)p(c)}$$
$p(c|a,b)$ известно из условия
$$p(c|a,b,d)=\frac{p(a,b,c,d)}{p(a,b,d)}=\frac{p(a,b,c,d)}{\sum_cp(a,b,c,d)}=\frac{p(d|c)p(c|a,b)p(a)p(b)}{\sum_cp(d|c)p(c|a,b)p(a)p(b)}=\frac{p(d|c)p(c|a,b)}{\sum_cp(d|c)p(c|a,b)}$$

```python
def a_x():
    return np.array(range(amin, amax+1))

def a(params, model=2):
    return uniform(amin, amax)
    
def b_x():
    return np.array(range(bmin, bmax+1))

def b(params, model=2):
    return uniform(bmin, bmax)  
```

```python
def c_x():
    return np.array(range(cmin, cmax+1))

def c_ab(a, b, params, model=2):
    c = np.zeros(cmax-cmin+1)
    b_range = range(b+1)
    n = a + b
    lam = a*p1+b*p2
    for i in xrange(n+1):
        if (model == 2):
            return c_ab_2(a, b, params)
        if (model == 1):
            return c_ab_1(a, b, params)     
    return c

def c_ab_1(a, b, params):
    c = np.zeros(cmax-cmin+1)
    a_max_local = min(a, a_max_value)
    b_max_local = min(b, b_max_value)
    for a_i in xrange(a_max_local+1):
        for b_i in xrange(b_max_local+1):
            n = a_i + b_i
            a_pr = bin(a, p1, a_i)
            b_pr = bin(b, p2, b_i)
            c[n] += a_pr*b_pr
    return c 

def c_ab_2(a, b, params):
    c = np.zeros(cmax-cmin+1)
    n = a + b
    lam = a*p1+b*p2
    for i in xrange(c_max_value+1):
        pr_c_n = poiss(i, lam)
        c[i] += pr_c_n
    return c

def c_a(a, params, model=2):
    c = np.zeros(cmax-cmin+1)
    pr_b = b(params, model)[0]
    for b_i in b_x():
        c += c_ab(a, b_i, params, model)
    c *= pr_b
    return c

def c_b(b, params, model=2):
    c = np.zeros(cmax-cmin+1)
    pr_a = a(params, model)[0]
    for a_i in a_x():
        c += c_ab(a_i, b, params, model)
    c *= pr_a
    return c

def c_b_lims(q, bmin, bmax, params, model=2):
    result = np.zeros(cmax-cmin+1)
    pr_b = b(params, model)[0]
    for b_i in range(bmin,bmax):
        result += c_b(b_i, params, model)
    result *= pr_b
    q.put(result)
        
def c(params, model=2):
    q = Queue()
    c = np.zeros(cmax-cmin+1)
    ranges = get_range(bmin, bmax+1, cpu_count*2)
    processes = []
    for i in xrange(cpu_count*2):
        current_range = ranges[i]
        bmin_val = current_range[0]
        bmax_val = current_range[1]
        p = Process(target=c_b_lims, args=(q, bmin_val, bmax_val, params, model,))
        p.start()
        processes.append(p)
        
    for p in processes:
        p.join()
        c += q.get()
    return c
    
```

```python
def d_x():
    return range(dmin, dmax+1)

def d_c(c_i, params, model=2):
    d = np.zeros(dmax+1)
    
    c_max_local = min(c_i, c_max_value)
    for f_stud in xrange(c_max_local+1):
        pr_d = bin(c_i, p3, f_stud)
        d_i = f_stud+c_i
        d[d_i] += pr_d
    return d

def d(params, model=2):
    c_distr = c(params, model)
    d = np.zeros(dmax+1)
    
    for c_i in xrange(c_max_value+1):
        pr_c_i = c_distr[c_i]
        f_stud_max_local = min(c_i, f_stud_max_value)
        for f_stud in xrange(f_stud_max_local+1):
            pr_d = bin(c_i, p3, f_stud)*pr_c_i
            d_i = f_stud+c_i
            d[d_i] += pr_d
    return d
```

```python
def c_d(d, params, model=2, c_computed=None):
    if (c_computed == None):
        c_computed = c(params, model) 
    c_d = np.zeros(cmax-cmin+1)
    for c_i in c_x()[d:]:
        pr_c = c_computed[c_i]
        f_stud = c_i - d
        pr_d_c = bin(c_i, p3, f_stud)
        pr_c_d = pr_d_c*pr_c
        c_d[c_i] += pr_c_d
    c_d_total = np.sum(c_d)
    c_d /= c_d_total
    return c_d

def c_abd(a,b,d, params, model=2):
    c_abd = np.zeros(cmax-cmin+1)
    c_ab_distr = c_ab(a,b,params, model)
    for c_i in c_x():
        d_c_distr = d_c(c_i, params, model)
        pr_d_c = d_c_distr[d]
        pr_c_ab = c_ab_distr[c_i]
        c_abd[c_i]= pr_d_c*pr_c_ab
    c_abd_total = np.sum(c_abd)
    return c_abd/c_abd_total
    
```

```python
def task_1(model=2):
    print "Оценка параметров априорных распределеий"
    a_probs = a(params, model)
    estimate_params("a", a_x(), a_probs)
    b_probs = b(params, model)
    estimate_params("b", b_x(), b_probs)
    c_probs = c(params, model)
    estimate_params("c", c_x(), c_probs)
    d_probs = d(params, model)
    estimate_params("d", d_x(), d_probs)
```

```python
def task_2_4(model=2):
    c_values = range(cmin, cmax+1)
    
    start = time.time()
    c_probs = c(params, model)
    print_time_estimation("c", time.time() - start)
    estimate_params("c", c_values, c_probs)
    plot_distr("p(c)", "c", c_values, c_probs)
    
    a_mean = int(mean(a_x(), a(params)))
    start = time.time()
    c_a_probs = c_a(a_mean, params, model)
    print_time_estimation("c|a", time.time() - start)
    estimate_params("c|a", c_values, c_a_probs)
    plot_distr("p(c|a)", "c", c_values, c_a_probs)
    
    b_mean = int(mean(b_x(), b(params)))
    start = time.time()
    c_b_probs = c_b(b_mean, params, model)
    print_time_estimation("c|b", time.time() - start)
    estimate_params("c|b", c_values, c_b_probs)
    plot_distr("p(c|b)", "c", c_values, c_b_probs)
    
    
    start = time.time()
    d_probs = d(params, model)
    print_time_estimation("d", time.time() - start)
    plot_distr("p(d)", "d", d_x(), d_probs)
    
    d_mean = int(mean(d_x(), d(params, model)))
    start = time.time()
    c_d_probs = c_d(d_mean, params, model)
    print_time_estimation("c|d", time.time() - start)
    estimate_params("c|d", c_values, c_d_probs)
    plot_distr("p(c|d)", "c", c_values, c_d_probs)
    
    
    start = time.time()
    c_a_b_probs = c_ab(a_mean, b_mean, params, model)
    print_time_estimation("c|ab", time.time() - start)
    estimate_params("c|ab", c_values, c_a_b_probs)
    plot_distr("p(c|ab)", "c", c_values, c_a_b_probs)
    
    start = time.time()
    c_a_b_d_probs = c_abd(a_mean, b_mean, d_mean, params, model)
    print_time_estimation("c|abd", time.time() - start)
    estimate_params("c|abd", c_values, c_a_b_d_probs)
    plot_distr("p(c|abd)", "c", c_values, c_a_b_d_probs)
```

```python
def task_3(model=2):
    c_values = c_x()
    a_devs = list()
    for a_i in a_x():
        c_a_probs = c_a(a_i, params, model)
        (a_mean, a_dev) = estimate_params("", c_values, c_a_probs, False)
        a_devs.append(a_dev)
        
    b_devs = list()
    for b_i in b_x():
        c_b_probs = c_b(b_i, params, model)
        (b_mean, b_dev) = estimate_params("", c_values, c_b_probs, False)
        b_devs.append(b_dev)
     
    
    d_devs = list()
    c_distr = c(params, model)
    for d_i in d_x():
        c_d_probs = c_d(d_i, params, model, c_distr)
        if (np.sum(c_d_probs) > 0.0):
            #skipping imposible situations
            #when all values in distribution is zero
            (d_mean, d_dev) = estimate_params("", c_values, c_d_probs, False)
            d_devs.append(d_dev)
        
    a_dev_min = np.min(np.array(a_devs))
    b_dev_min = np.min(np.array(b_devs))
    d_dev_max = np.max(np.array(d_devs))
    
    print "min D(c|a)={0}, min D(c|b)={1}, max D(c|d)={2}".format(a_dev_min, b_dev_min, d_dev_max)
    print "d влияет на дисперсию больше всех = ", d_dev_max<a_dev_min and d_dev_max<b_dev_min
    
    first_x = list()
    first_y = list()
    second_x = list()
    second_y = list()
    for b_i in b_x():
        for a_i in a_x():
            if (b_devs[b_i-bmin]<a_devs[a_i-amin]):
                first_x.append(b_i)
                first_y.append(a_i)
            else:
                second_x.append(b_i)
                second_y.append(a_i)
                
    plt.plot(first_x, first_y, 'o', color='blue')
    plt.plot(second_x, second_y, 'o', color='red')
    plt.show()
                
```

```python
def perform_tasks(model=2):
    task_1(model)
    task_2_4(model)
    task_3(model)
```
